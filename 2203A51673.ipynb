```python
import math
def sin(x):
    return math.sin(x)
def log(x):
    return math.log(x)
def log2(x):
    return math.log2(x)
def exp(x):
    return math.exp(x)
print("sin(0):", sin(0))
print("log(1):", log(1))
print("log2(2):", log2(2))
print("exp(1):", exp(1))
```

```python

import matplotlib.pyplot as plt
import pandas as pd


people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]


df = pd.DataFrame({
    'People': people,
    'Age': age,
    'Height': height,
    'Weight': weight
})
plt.figure(figsize=(10,6))
plt.subplot(1, 3, 1)
plt.bar(df['People'], df['Age'], color='blue')
plt.title('Age')
plt.xlabel('People')
plt.ylabel('Age')
plt.subplot(1, 3, 2)
plt.bar(df['People'], df['Height'], color='green')
plt.title('Height')
plt.xlabel('People')
plt.ylabel('Height')
plt.subplot(1, 3, 3)
plt.bar(df['People'], df['Weight'], color='red')
plt.title('Weight')
plt.xlabel('People')
plt.ylabel('Weight')
plt.tight_layout()
plt.show()
```
